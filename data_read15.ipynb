```python
df = pd.read_csv(r"building data/2zone_Supermarket.csv", encoding="gbk")
df.rename(columns={'Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)': 'outdoor temperature'}, inplace=True)
df
```

```python
from pathlib import Path
import os
```

```python
folder = Path(r"integrate2023")
csv_files = [p for p in folder.iterdir()]
key_col = "time"
dfs = []
for f in csv_files:
    # df = pd.read_csv(f, encoding='utf-8', encoding_errors='ignore')
    df = pd.read_csv(f, encoding='utf-8')
    key = df.columns[0]
    stem = f.stem
    df.set_index(key).add_suffix(f"_{stem}").pipe(dfs.append)

merged = pd.concat(dfs, axis=1)
merged.reset_index(inplace=True)

print(merged.shape)
merged['Time'] = pd.to_datetime(merged['Time'], format="%Y/%m/%d %H:%M")
min15_exact = merged[merged['Time'].dt.minute.isin([0, 15, 30, 45])].copy()
min15_exact = min15_exact.sort_values(by='Time')
min15_exact
# merged.set_index('Time')
# min15_exact = merged[merged.index.minute == 0 or merged.index.minute == 15 or merged.index.minute == 30 or merged.index.minute == 45]
```

```python
min15_exact.isna().stack()[lambda x: x].index.tolist()
min15_exact = min15_exact.interpolate(method='linear')
min15_exact.isna().sum()
```

```python
pv_file = Path(r"/Users/liqi/Library/CloudStorage/OneDrive-TheUniversityofHongKong-Connect/DRMPC with Prob Forecast/datasets/solar power Hong Kong rooftop/Time series dataset/PV generation dataset/PV stations without panel level optimizer/Site level dataset/CYT (Station 1).csv")
pv = pd.read_csv(pv_file)
pv["Time"] = pd.to_datetime(pv["Time"], format="%Y-%m-%d %H:%M:%S")
pv_select = pv[pv["Time"].dt.year == 2023]
pv_select
```

```python
pv_select_indexed = pv_select.set_index('Time')
pv_15 = pv_select_indexed.resample('15T').interpolate(method='linear')

pv_15 = pv_15.reset_index()
last_row = pv_15.iloc[-1].copy()
for minute in [15, 30, 45]:
    new_row = last_row.copy()
    new_row['Time'] = pd.Timestamp('2023-12-31 23:00:00') + pd.Timedelta(minutes=minute)
    pv_15 = pd.concat([pv_15, pd.DataFrame([new_row])], ignore_index=True)

pv_15 = pv_15.sort_values('Time').reset_index(drop=True)
pv_15
```

```python
pv_15.isna().sum()
```

```python
new = pd.merge(
    pv_15, min15_exact,
    on= 'Time',
    how = "outer"
).sort_values("Time").reset_index(drop=True)
new
```

```python
new.isna().sum()
```

```python
# col_to_fill = ["RH (%)_Relative Humidity_2022", "Wind Speed (m/s)_Wind_2022", "Wind Direction (degree)_Wind_2022",
#               "Temp (Degree Celsius)_Temperature_2022", "Rainfall(mm)_Rainfall_2022",
#               "Vis (km)_Visibility_2022", "Irradiance (W/m2)_Irradiance_2022", "SLP (hPa)_Sea Level Pressure_2022"]
col_to_fill = ["RH (%)_Relative Humidity_2023", "Wind Speed (m/s)_Wind_2023", "Wind Direction (degree)_Wind_2023",
              "Temp (Degree Celsius)_Temperature_2023", "Rainfall(mm)_Rainfall_2023",
              "Vis (km)_Visibility_2023", "Irradiance (W/m2)_Irradiance_2023", "SLP (hPa)_Sea Level Pressure_2023"]
max_iter = 7
max_iter = 7

for col in col_to_fill:
    for iter in range(max_iter):
        new[col] = new[col].fillna(new[col].shift(96))
        if new[col].isna().sum() == 0:
            break

new.isna().sum()
```

```python
data = new.copy()
data["month"] = data["Time"].dt.month
data["day"] = data["Time"].dt.day
data["hour"] = data["Time"].dt.hour
data["min"] = data["Time"].dt.minute

cols = data.columns.tolist()
# new_cols = ['month', 'day', 'hour', 'min', 'generation(kWh)', 'RH (%)_Relative Humidity_2022', 
#             'Wind Speed (m/s)_Wind_2022', 'Wind Direction (degree)_Wind_2022', 'Temp (Degree Celsius)_Temperature_2022',
#             'Rainfall(mm)_Rainfall_2022', 'Vis (km)_Visibility_2022', 'Irradiance (W/m2)_Irradiance_2022',
#             'SLP (hPa)_Sea Level Pressure_2022']
new_cols = ['month', 'day', 'hour', 'min', 'generation(kWh)', 'RH (%)_Relative Humidity_2023', 
            'Wind Speed (m/s)_Wind_2023', 'Wind Direction (degree)_Wind_2023', 'Temp (Degree Celsius)_Temperature_2023',
            'Rainfall(mm)_Rainfall_2023', 'Vis (km)_Visibility_2023', 'Irradiance (W/m2)_Irradiance_2023',
            'SLP (hPa)_Sea Level Pressure_2023']
final_data = data[new_cols]
final_data
```
